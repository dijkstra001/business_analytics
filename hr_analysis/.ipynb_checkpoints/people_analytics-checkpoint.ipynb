```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sbn
```

```python
dataset = pd.read_csv('dataset.csv')
dataset.head()
```

```python
# Change type of columns for a vector (only categorical columns):

dataset['Attrition'] = pd.Categorical(dataset['Attrition'])
dataset['BusinessTravel'] = pd.Categorical(dataset['BusinessTravel'])
dataset['Department'] = pd.Categorical(dataset['Department'])
dataset['Education'] = pd.Categorical(dataset['Education'])
dataset['EducationField'] = pd.Categorical(dataset['EducationField'])
dataset['Employee Source'] = pd.Categorical(dataset['Employee Source'])
dataset['EnvironmentSatisfaction'] = pd.Categorical(dataset['EnvironmentSatisfaction'])
dataset['Gender'] = pd.Categorical(dataset['Gender'])
dataset['JobInvolvement'] = pd.Categorical(dataset['JobInvolvement'])
dataset['JobLevel'] = pd.Categorical(dataset['JobLevel'])
dataset['JobRole'] = pd.Categorical(dataset['JobRole'])
dataset['JobSatisfaction'] = pd.Categorical(dataset['JobSatisfaction'])
dataset['MaritalStatus'] = pd.Categorical(dataset['MaritalStatus'])
dataset['OverTime'] = pd.Categorical(dataset['OverTime'])
dataset['PerformanceRating'] = pd.Categorical(dataset['PerformanceRating'])
dataset['RelationshipSatisfaction'] = pd.Categorical(dataset['RelationshipSatisfaction'])
dataset['StockOptionLevel'] = pd.Categorical(dataset['StockOptionLevel'])
dataset['WorkLifeBalance'] = pd.Categorical(dataset['WorkLifeBalance'])
```

```python
# Atribute Engineering:

# Create column to visualize years of experience:
dataset['PriorYearOfExperience'] = dataset['TotalWorkingYears'] - dataset['YearsAtCompany']

# Create column to visualize average tenure:
dataset['AverageTenure'] = dataset['PriorYearOfExperience'] / dataset['NumCompaniesWorked']
dataset['AverageTenure'] = dataset['AverageTenure'].replace([np.inf], 0.0)

# Create dataset to visualize dismissed employee:
dataset_dism_employee = dataset[dataset['Attrition'] != 'Termination']
```
